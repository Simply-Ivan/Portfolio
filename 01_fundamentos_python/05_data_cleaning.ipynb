### DataFrame Merges

```python
df1 = DataFrame({'key': ['b','b','a','c','a','a','b'],
                 'data1': range(7)})
df1
```

```python
df2 = DataFrame({'key': ['a','b','d'],
                 'data2': range(3)})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3 = DataFrame({'lkey': ['b','b','a','c','a','a','b'],
                 'data1': range(7)})
df3
```

```python
df4 = DataFrame({'rkey': ['a','b','d'],
                 'data2': range(3)})
df4
```

```python
pd.merge(df3,df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df1,df2, how='outer')
```

```python
df1 = DataFrame({'key': ['b','b','a','c','a','b'],
                 'data1': range(6)})
df1
```

```python
df2 = DataFrame({'key': ['a','b','a','b','d'],
                 'data2': range(5)})
df2
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2, how='inner')
```

```python
left = DataFrame({'key1': ['foo', 'foo', 'bar'],
        'key2': ['one', 'two', 'one'],
        'lval': [1, 2, 3]})
left
```

```python
right = DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
        'key2': ['one', 'one', 'one', 'two'],
        'rval': [4, 5, 6, 7]})
right
```

```python
pd.merge(left,right, on=['key1','key2'], how='outer')
```

```python
pd.merge(left,right, on='key1', how='inner')
```

```python
pd.merge(left,right, on='key1', suffixes=('_left','_right'))
```

| Argumento      | Descripción |
|----------------|-------------|
| `left`         | DataFrame que se va a unir por el lado izquierdo |
| `right`        | DataFrame que se va a unir por el lado derecho |
| `how`          | Tipo de unión: `'inner'`, `'outer'`, `'left'` o `'right'`. Por defecto es `'inner'` |
| `on`           | Nombre(s) de columna(s) para realizar la unión. Deben estar en ambos DataFrames. Si no se especifica y no se dan otras claves de unión, se usará la intersección de los nombres de columnas de `left` y `right` |
| `left_on`      | Nombre(s) de columna(s) del DataFrame izquierdo a usar como clave de unión |
| `right_on`     | Igual que `left_on`, pero para el DataFrame derecho |
| `left_index`   | Usa el índice del DataFrame izquierdo como clave de unión (o múltiples si es un MultiIndex) |
| `right_index`  | Igual que `left_index`, pero para el DataFrame derecho |
| `sort`         | Ordena el resultado de la unión por las claves de unión de forma lexicográfica. Es `True` por defecto. Puedes desactivarlo para mejorar el rendimiento en conjuntos de datos grandes |
| `suffixes`     | Tupla de cadenas para añadir a los nombres de columna en caso de que haya nombres duplicados. Por defecto es `('_x', '_y')`. Por ejemplo, si ambas tablas tienen una columna `data`, en el resultado aparecerán como `data_x` y `data_y` |
| `copy`         | Si es `False`, evita copiar datos en el resultado en ciertos casos especiales. Por defecto, siempre copia |

### Merging on Index

```python
left1 = DataFrame({'key': ['a','b','a','a','b','c'],
                   'value': range(6)})
left1
```

```python
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a','b'])
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
lefth = DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
                   'key2': [2000, 2001, 2002, 2001, 2002],
                   'data': np.arange(5.)})
lefth
```

```python
righth = DataFrame(np.arange(12).reshape((6, 2)),
                   index=[['Nevada', 'Nevada', 'Ohio', 'Ohio', 'Ohio', 'Ohio'],
                          [2001, 2000, 2000, 2000, 2001, 2002]],
                   columns=['event1', 'event2'])
righth
```

```python
pd.merge(lefth, righth, left_on=['key1','key2'], right_index=True)
```

```python
pd.merge(lefth, righth, left_on=['key1','key2'], right_index=True, how='outer')
```

```python
left2 = DataFrame([[1., 2.], [3., 4.], [5., 6.]], index=['a', 'c', 'e'],
                   columns=['Ohio', 'Nevada'])
left2
```

```python
right2 = DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                   index=['b', 'c', 'd', 'e'], columns=['Missouri', 'Alabama'])
right2
```

```python
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
left2.join(right2, how='outer')
```

```python
left1.join(right1, on='key')
```

```python
another = DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                     index=['a', 'c', 'e', 'f'], columns=['New York', 'Oregon'])
another
```

```python
left2.join([right2, another])
```

```python
left2.join([right2, another], how='outer')
```

### Transforming Data Using a Function or Mapping

```python
data = DataFrame({'food': ['bacon','pulled pork', 'bacon', 'Pastrami',
                           'corned beef', 'Bacon', 'pastrami', 'honey ham',
                           'nova lox'],
                  'ounces': [4,3,12,6,7.5,8,3,5,6]})
data
```

```python
meat_to_animal = {
'bacon': 'pig',
'pulled pork': 'pig',
'pastrami': 'cow',
'corned beef': 'cow',
'honey ham': 'pig',
'nova lox': 'salmon'
}
```

```python
data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
data
```

```python
data['food'].map(lambda x: meat_to_animal[x.lower()])
```
